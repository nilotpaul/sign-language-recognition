```python
import boto3
from boto3.exceptions import Boto3Error
from decouple import config

s3 = boto3.client('s3')

bucket_name = config('AWS_BUCKET_NAME')
words = []
picture_urls = []

def get_words(folder_name):
    idx = 0

    try:
       res = s3.list_objects_v2(Bucket=bucket_name, Prefix=folder_name, Delimiter='/')
    except Boto3Error as e:
        print(f'Get image Error: {e}')
    except Exception as e:
        print(f'Get image Error: {e}')
    
    for prefix in res['CommonPrefixes']:
        if (prefix):
            words.append(prefix['Prefix'].split('/')[1])
            idx += 1
    
    return words

def get_pictures(folder_name):
    idx = 0

    try:
       res = s3.list_objects_v2(Bucket=bucket_name, Prefix=folder_name, Delimiter='/')
    except Boto3Error as e:
        print(f'Get image Error: {e}')
    except Exception as e:
        print(f'Get image Error: {e}')
    
    for content in res['Contents']:
        if (content):
            picture_urls.append(res['Contents'][idx]['Key'])
            idx += 1
    
    return picture_urls
```

```python
words = get_words('pictures/')

total_words = len(words)

print(total_words)
```

```python
words = get_words('pictures/')

pics = []

for word in words[0:2]:
    pic = get_pictures(f'pictures/{word}/')

    pics.append({
        'word': word,
        'image_url': pic
    })

print(pics)
```
